```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ChurN_Modelling.csv')
df
```

```python
x = df.iloc[:, 3:13] # 1st 3 cols are not required
y = df.iloc[:, 13]
```

```python
geography = pd.get_dummies(x["Geography"],drop_first=True)
gender = pd.get_dummies(x.Gender, drop_first=True)
```

```python
x = pd.concat([x, geography, gender], axis=1)
x
```

```python
x = x.drop(['Geography', 'Gender'], axis = 1)
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
# now data is in same scale
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
# from keras.layers import LeakyReLU,PReLLU,ELU
from keras.layers import Dropout
```

```python
classifer = Sequential()
```

```python
classifer.add(Dense(units=10, kernel_initializer='he_normal', activation='relu', input_dim = 11))
classifer.add(Dropout(0.3))
classifer.add(Dense(units=20, kernel_initializer='he_normal', activation='relu'))
classifer.add(Dropout(0.4))
classifer.add(Dense(units=15, kernel_initializer='he_normal', activation='relu'))
classifer.add(Dropout(0.2))
classifer.add(Dense(units=1, kernel_initializer='glorot_uniform', activation='sigmoid'))
# 1st layer - input - 11 nuerons coz 11 columns i.e. 11 features
# 2nd layer - hidden - 10 nuerons
# 3rd layer - hidden - 20 nuerons
# 4th layer - hidden - 15 nuerons
# 5th layer - output - 1 neuron
```

```python
classifer.summary()
```

```python
classifer.compile(optimizer='Adamax', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model_history = classifer.fit(x_train,y_train,validation_split=.33,batch_size=10,nb_epoch=100)
```

```python
y_pred = classifer.predict(x_test)
y_pred = y_pred > .5
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred)
score
```
